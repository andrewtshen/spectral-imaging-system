```python
import PIL
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
img = PIL.Image.open("park.jpg")
img_array = np.array(img)
plt.imshow(img_array)
```

```python
half = int(4096//2)
rgb_channel = img_array[:,0:half,:3]
ir_image = img_array[:,half:,:3]
rgb_channel.shape
```

```python
plt.imshow(rgb_channel)
```

```python
plt.imshow(ir_image, cmap="gray")
```

```python
def get_rescaled_intensities(ir_image):
    ir_image_copy = np.copy(ir_image)
    intensity = ir_image_copy[:,:,0].astype(np.uint16) + ir_image_copy[:,:,1].astype(np.uint16) + ir_image_copy[:,:,2].astype(np.uint16) #new max value is between 0 and 255*3
    rescaled_intensity = np.copy(intensity).astype(float) / intensity.max()
    return rescaled_intensity
```

```python
def display_image(img, ir_values, value): #between 1 and 1.2
    ir_values_copy = np.copy(ir_values)
    rescaled_intensities = get_rescaled_intensities(ir_values_copy)
    img[:, :, 0] = img[:, :, 0] * rescaled_intensities.astype(float) * value  # scaling the r channel
    img[:, :, 1] = img[:, :, 1] * rescaled_intensities.astype(float) * value # scaling the g channel
    img[:, :, 2] = img[:, :, 2] # keep the b channel the same
    plt.imshow(img)
```

```python
display_image(np.copy(rgb_channel), np.copy(ir_image), 1.1)
```
