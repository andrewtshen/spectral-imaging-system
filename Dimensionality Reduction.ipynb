```python
from sklearn.decomposition import PCA
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
# sample_im = np.load("example/2021-11-15_22-52-14.jpg.npy")
sample_im = np.load("ALIGNED_DATASET/stacked/2021-11-15_23-07-37_combined.npy")
```

```python
def reduce_dims(im):
    flat_sample_im = im.reshape(-1, 4)
    pca = PCA(n_components=3)
    # pca.fit(flat_sample_im)
    # reshaped_im = sample_im @ pca.components_.T
    new_sample_im = pca.fit_transform(flat_sample_im)
    transformed_im = new_sample_im.reshape((im.shape[0], im.shape[1], 3))
    transformed_im = np.interp(transformed_im, (transformed_im.min(), transformed_im.max()), (0, 255))/256.
    return transformed_im
```

```python
plt.imshow(reduce_dims(sample_im))
```
